```python
area =[600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
plt.plot(area,price)
```

```python
#change size of plot
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
#adding title
plt.title("area vs price")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
#adding title
plt.title("area vs price")
#labels
plt.xlabel("area in sq.ft")
plt.ylabel("price in thousands")
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
#adding title
plt.title("area vs price")
#labels
plt.xlabel("area in sq.ft")
plt.ylabel("price in thousands")
#change limit/range
plt.xlim((400,2000))
plt.ylim((5,50))
#tick rotation
plt.xticks(rotation=90)# in degrees
plt.yticks(rotation=45)# in degrees
plt.show()
```

```python
# multiple plots
x= np.arange(1,20)
y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.plot(x,y1,color="yellow")
plt.plot(x,y2,color="red")
plt.show()
```

```python
plt.plot(x,y1,color="yellow")
plt.plot(x,y2,color="red")
plt.plot(x,x,color="green")
#legend
plt.legend(labels=['sqrt','log','linear'])
plt.show()
```

# Histogram

```python
df = pd.read_csv("cars.csv")
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#plt.figure(figsize=(7,7))
plt.xlabel("width")
plt.ylabel("frequency")
#how many bars are displayed using bins
plt.hist(df['width'],bins=10)
plt.show()
```

```python
#using pandas
df['width'].hist()
plt.show()
```

```python
df['width'].hist(bins=15)
plt.show()
```

```python
#plt axvline
plt.hist(df['width'])
width_mean=df['width'].mean()
width_median=df['width'].median()
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='yellow')
plt.show()
```

```python
df['width'].hist()
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='yellow')
plt.show()
```

# Bar plots

```python
#kind = bar ,barh
# display make wise count
df['make'].value_counts()
```

```python
df['make'].value_counts().plot(kind='bar')
plt.xlabel("make")
plt.ylabel("count")
plt.show()
```

```python
df['make'].value_counts().plot(kind='barh')
plt.xlabel("make")
plt.ylabel("count")
plt.show()
```
