```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
# increase size of the plot
# plt.figure(figsize=(width, height))
plt.figure(figsize=(7,7))
plt.plot(area,price)
#add information of the graph
#add title
plt.title("Area vs Price")
#add label for x and y
plt.xlabel("Area")
plt.ylabel("Price")
#adding limit to the plot
plt.xlim((400,2000))
plt.ylim((5,50))
plt.show()
```

```python
plt.plot(area,price)
#add information of the graph
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
#rotate ticks
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.show()
```

```python
x = np.arange(1,20)
y1= np.sqrt(x)
y2= np.log(x)
plt.figure(figsize=(7,7))
plt.plot(x,y1,'bx-')
plt.plot(x,y2,'go-')
plt.plot(x,x,'r')
plt.legend(labels=['sqrt','log','linear'])
plt.show()
```

```python
df = pd.read_csv("cars.csv")
```

```python
df.info()
```

```python
df
```

```python
plt.figure(figsize=(7,7))
plt.title("width frequency distribution")
plt.ylabel("frequency")
plt.xlabel("width")
#bins are no of bars in histogram
plt.hist(df["width"],bins=10)
plt.show()
#graph is normally distributed
```

```python
df["width"].hist(bins=15)
plt.show()
```

```python
width_mean=df["width"].mean()
width_median=df["width"].median()
```

```python
plt.figure(figsize=(7,7))
plt.hist(df["width"],bins=10)
plt.axvline(width_mean, color="red")
plt.axvline(width_median, color="green")
plt.title("width frequency distribution")
plt.ylabel("frequency")
plt.xlabel("width")
plt.show()
```

```python
plt.figure(figsize=(7,7))
df["width"].hist(bins=10)
plt.axvline(width_mean, color="red")
plt.axvline(width_median, color="green")
plt.title("width frequency distribution")
plt.ylabel("frequency")
plt.xlabel("width")
plt.show()
```
